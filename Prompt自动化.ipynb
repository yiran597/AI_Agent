```python
import json
import os
from docx import Document

# 文件夹路径
word_folder = "/content/drive/MyDrive/lilysilk/自研AI Agent/prompts"

# 解析 Word 文件内容（支持标题 + 段落）
def extract_text_from_docx(file_path):
    doc = Document(file_path)
    text_data = {}
    current_section = None

    for para in doc.paragraphs:
        text = para.text.strip()
        if not text:
            continue

        # 确保 `para.style` 存在，避免 `NoneType` 错误
        para_style = getattr(para.style, "name", "")

        # 识别 Word 标题（适配不同格式）
        if para_style.startswith("Heading") or text in ["目标", "背景", "技能大框架", "关键技能", "工作方式", "输出格式", "回答风格", "回答限制"]:
            current_section = text  # 记录当前标题
            text_data[current_section] = []
        elif current_section:
            text_data[current_section].append(text)

    return text_data

# 遍历所有 Word 文件并转换为 JSON
def convert_word_to_json(folder_path):
    all_prompts = {}
    for file in os.listdir(folder_path):
        if file.endswith(".docx"):
            file_path = os.path.join(folder_path, file)
            prompt_key = file.replace(".docx", "").lower().replace(" ", "_")  # 生成 JSON 键名
            extracted_data = extract_text_from_docx(file_path)
            if extracted_data:  # 确保不是空的
                all_prompts[prompt_key] = extracted_data

    # 存储 JSON
    json_path = "/content/drive/MyDrive/lilysilk/自研AI Agent/prompts_fixed.json"
    with open(json_path, "w", encoding="utf-8") as json_file:
        json.dump(all_prompts, json_file, ensure_ascii=False, indent=4)

    print(f"JSON 文件已保存至: {json_path}")

# 运行转换
convert_word_to_json(word_folder)
```

动态填充 Prompt：
动态替换 Prompt 里的变量

```python
def generate_prompt(prompt_type, **kwargs):
    """获取 Prompt 并填充变量"""
    prompt = get_prompt(prompt_type)
    if isinstance(prompt, dict):
        prompt_text = "\n".join([f"{key}: {', '.join(value) if isinstance(value, list) else value}" for key, value in prompt.items()])
        return prompt_text.format(**kwargs)
    return "未知的 Prompt 类型"

# 示例调用（自动填充角色）
print(generate_prompt("商业数据分析师prompt", role="数据分析师", focus="业务增长"))
```

智能分类系统：根据 用户输入的文本内容 自动匹配 最合适的 Prompt，然后返回对应的 Prompt 内容。

```python
# 创建 Prompt 关键词映射
## 手动定义 关键词，确保不同的用户输入 能匹配到正确的 Prompt
# 关键词映射表
prompt_keywords = {
    "商业数据分析师prompt": ["数据分析", "数据挖掘", "统计建模", "数据驱动"],
    "商业智能分析师prompt": ["商业智能", "BI", "数据可视化", "商业洞察"],
    "数字策略师prompt": ["数字营销", "市场策略", "品牌推广", "增长策略"],
    "数据分析师prompt": ["数据预处理", "数据可视化", "SQL", "机器学习"],
    "数据可视化工程师prompt": ["可视化", "BI 报表", "图表", "交互式数据"],
    "电商运营师prompt": ["电商", "市场推广", "供应链", "用户增长"]
}
```

```python
# 实现匹配逻辑
## 使用 文本相似度计算（模糊匹配）
from difflib import get_close_matches

def detect_prompt_type(user_input):
    """基于用户输入，匹配最佳 Prompt 类型"""
    for prompt, keywords in prompt_keywords.items():
        matches = get_close_matches(user_input, keywords, n=1, cutoff=0.5)  # 计算相似度
        if matches:
            return prompt  # 返回匹配的 Prompt 名称
    return "default_prompt"  # 默认返回

# 示例调用
user_query = "我想做商业智能数据分析"
selected_prompt = detect_prompt_type(user_query)
print(f"匹配到的 Prompt: {selected_prompt}")
```

```python
# 返回最佳匹配的 Prompt
def get_prompt(prompt_type):
    """获取匹配的 Prompt 内容"""
    return prompts.get(prompt_type, "未知的 Prompt 类型")

# 自动匹配并获取 Prompt
matched_prompt = detect_prompt_type("我要做商业智能分析")
prompt_content = get_prompt(matched_prompt)
print(prompt_content)
```

***实现交互***

使用 NLTK 过滤掉无关词

提取句子中的核心关键词

使用工具：BERT。模型：sentence-transformers/paraphrase-MiniLM-L6-v2

优化：
- 提高 BERT 相似度阈值（从 0.5 提高到 0.7）。
- 增强 Prompt 关键词集合（确保比如 数据分析 和 电商 不会因 单个关键词相似 而误匹配）。
- Top-N 策略（如果 第一匹配不够高，会自动考虑 第二匹配）。

```python
import json
import torch
from sentence_transformers import SentenceTransformer, util

# **加载轻量级 BERT 模型**
model = SentenceTransformer("sentence-transformers/paraphrase-MiniLM-L6-v2", device="cpu")

# **加载 JSON Prompt 数据**
json_path = "/content/drive/MyDrive/lilysilk/自研AI Agent/prompts_fixed.json"
with open(json_path, "r", encoding="utf-8") as f:
    prompts = json.load(f)

# **改进 Prompt 关键词**
prompt_keywords = {
    "商业数据分析师prompt": ["数据分析", "数据挖掘", "统计建模", "数据驱动", "分析报告", "数据趋势", "数据可视化"],
    "商业智能分析师prompt": ["商业智能", "BI", "数据可视化", "商业洞察", "数据监测", "数据平台", "数据整合"],
    "数字策略师prompt": ["数字营销", "市场策略", "品牌推广", "增长策略", "用户增长", "营销方案", "市场分析"],
    "数据分析师prompt": ["数据预处理", "数据可视化", "SQL", "机器学习", "数据建模", "商业数据", "用户增长分析", "数据挖掘"],
    "数据可视化工程师prompt": ["数据可视化", "BI 报表", "图表", "交互式数据", "数据监测", "数据分析"],
    "电商运营师prompt": ["电商", "市场推广", "电商策略", "供应链优化", "用户增长", "销售数据", "营销增长", "电商增长策略"]
}

# **计算所有 Prompt 关键词的 BERT 嵌入**
corpus = [" ".join(keywords) for keywords in prompt_keywords.values()]
prompt_embeddings = model.encode(corpus, convert_to_tensor=True)

def detect_prompt_type_bert(user_input):
    """使用 BERT 语义匹配 Prompt，并优化误匹配问题"""
    user_embedding = model.encode(user_input, convert_to_tensor=True)
    similarities = util.pytorch_cos_sim(user_embedding, prompt_embeddings)[0]  # 计算相似度
    sorted_indices = torch.argsort(similarities, descending=True)  # 排序获取 Top-N 结果

    best_match_index = sorted_indices[0].item()  # 最高相似度
    best_prompt = list(prompt_keywords.keys())[best_match_index]

    # **如果相似度太低，检查第二名**
    if similarities[best_match_index] < 0.7 and len(sorted_indices) > 1:
        second_match_index = sorted_indices[1].item()
        second_best_prompt = list(prompt_keywords.keys())[second_match_index]
        print(f"\n 请注意，置信度较低，可能的备选 Prompt: {second_best_prompt} (相似度: {similarities[second_match_index]:.2f})")

        # **如果次高的 Prompt 置信度更接近用户需求，就用次高**
        if similarities[second_match_index] > 0.65:
            best_prompt = second_best_prompt

    return best_prompt if similarities[best_match_index] > 0.7 else "default_prompt"

# **获取 Prompt 详细内容**
def get_prompt(prompt_type):
    return prompts.get(prompt_type, "未知的 Prompt 类型")

# **交互式输入**
while True:
    user_query = input("\n 请输入你的需求（输入 '退出' 结束）：")
    if user_query.lower() == "退出":
        break

    matched_prompt = detect_prompt_type_bert(user_query)

    # **询问用户是否确认**
    while True:
        confirm = input(f"\n **匹配到的 Prompt:** {matched_prompt} \n 这符合你的角色和提问吗？(是/否)：").strip().lower()

        if confirm == "是":
            print("\n **请按照以下格式提供你的具体需求：**")
            print(" **目标:** 你希望这个分析的目标是什么？")
            print(" **背景:** 你可以提供哪些已有的数据或上下文？")
            print(" **具体需求:** 你希望系统提供什么样的结果？\n")
            break  # 继续执行逻辑

        elif confirm == "否":
            print("\n **请重新输入你的需求！**")
            break  # 重新输入

    # **获取 Prompt 详情**
    prompt_content = get_prompt(matched_prompt)
    print("\n **完整 Prompt 内容:**")
    print(json.dumps(prompt_content, indent=4, ensure_ascii=False))
```
